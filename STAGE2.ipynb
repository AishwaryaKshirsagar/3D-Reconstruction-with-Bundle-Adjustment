# Keypoint maching

```python
def correspondance_pipeline(corr_path, imgs_path, method='SIFT', lowe_ratio=0.7, lowe_min_dist=150):
    imgs = get_all_imgs_paths_in_folder(imgs_path)
    generate_img_correspondances(img_paths=imgs, method=method, num_keypoints=2000, output_folder=corr_path, lowe_ratio=0.7, lowe_min_dist=150)

    for file in os.listdir(corr_path):
            if os.path.isfile(os.path.join(corr_path, file)):
                idx1 = file.split('_')[0]
                idx2 = file.split('_')[1].replace(".txt", '')
                correspondences = load_correspondences(os.path.join(corr_path, file))
                
                img1_path = imgs_path+idx1+".jpg"
                img2_path = imgs_path+idx2+".jpg"
                print("BEFORE CLEANING")
                visualize_correspondences(img1_path, img2_path, correspondences)
                print("AFTER CLEANING")
                correspondences = clean_correspondances(correspondences)
                save_correspondences(file_path=os.path.join(corr_path, file), correspondences=correspondences)
                visualize_correspondences(img1_path, img2_path, correspondences)
```

```python
corr_path = "milk/mini_masked_corrs"
imgs_path = "milk/mini_masked_images/"

correspondance_pipeline(corr_path, imgs_path)
```

```python
corr_path = "milk/corrs"
imgs_path = "milk/images/"

correspondance_pipeline(corr_path, imgs_path)
```

```python
corr_path = "boot/corrs"
imgs_path = "boot/images/"
imgs = get_all_imgs_paths_in_folder(imgs_path)

correspondance_pipeline(corr_path, imgs_path)
```

```python
corr_path = "milk/masked_corrs"
imgs_path = "milk/masked_images/"

#We can relax the lowe conditions a bit because we have segmented the images
correspondance_pipeline(corr_path, imgs_path, lowe_ratio=0.7, lowe_min_dist=100)
```
